```python
from pathlib import Path
from dotenv import dotenv_values

ENV_VARS = dotenv_values("../.env")
DPATH_DATA = Path(ENV_VARS["DPATH_FL_DATA_LATEST"])

DATASET_MAP = {
    "ppmi": "PPMI",
    "calgary": "Calgary",
    "qpn": "QPN",
    "adni": "ADNI",
    "pad": "PREVENT-AD",
}

DATASETS_PER_TASK = {
    "Age task": DATASET_MAP.keys(),
    "Cognitive decline task": DATASET_MAP.keys(),
    "Diagnosis task": [key for key in DATASET_MAP.keys() if key != "pad"],
}
```

```python
import pandas as pd
from fl_pd.custom_dataset import training_plan_factory


def load_data(tag: str) -> pd.DataFrame:
    training_plan_class = training_plan_factory(tag)
    dfs = {}
    for dataset_name in DATASET_MAP.keys():
        if dataset_name.lower() == "cogtips":
            continue
        try:
            dataset = training_plan_class.dataset_factory(target=tag, i_split=0)
            dataset.path = DPATH_DATA / dataset_name.lower()
            dataset.read()
            dfs[dataset_name] = dataset.df_after_transforms
        except KeyError:
            continue

    df = pd.concat(dfs, axis="index", names=["dataset", "participant_id", "session_id"])
    df = df.reset_index(level=["participant_id", "session_id"])  # , drop=True)
    df = df.reset_index()
    return df


df_diag = load_data("nb:Diagnosis")
df_cog_decline = load_data("fl:cognitive_decline_status")
df_age = load_data("nb:Age")
```

```python
def load_harmonized(dataset):
    fpath = DPATH_DATA / dataset / "tabular" / "harmonized.tsv"
    df = pd.read_csv(
        fpath,
        sep="\t",
        index_col=["nb:ParticipantID", "nb:SessionID"],
        dtype={"nb:ParticipantID": str, "nb:SessionID": str},
    )
    return df


df_harmonized = {}
for dataset in DATASET_MAP.keys():
    df_harmonized[dataset] = load_harmonized(dataset)
```

```python
from rich.table import Table

COL_AGE = "nb:Age"
COL_SEX = "nb:Sex"
COL_DIAGNOSIS = "nb:Diagnosis"

TERMURL_MALE = "snomed:248153007"
TERMURL_FEMALE = "snomed:248152002"
TERMURL_HEALTHY_CONTROL = "ncit:C94342"


def get_row_data(df):

    def ratio(series):
        count = sum(series)
        return f"{count:.0f} ({count / len(series) * 100:.1f}%)"

    sample_size = len(df)

    age = f"{df[COL_AGE].mean():.1f} ± {df[COL_AGE].std():.1f}"

    male = ratio(df[COL_SEX] == TERMURL_MALE)

    female = ratio(df[COL_SEX] == TERMURL_FEMALE)

    # healthy_control = ratio(df[COL_DIAGNOSIS] == TERMURL_HEALTHY_CONTROL)

    # case = ratio(df[COL_DIAGNOSIS] != TERMURL_HEALTHY_CONTROL)

    # sanity checks
    if not sample_size == sum((df[COL_SEX] == TERMURL_MALE)) + sum(
        (df[COL_SEX] == TERMURL_FEMALE)
    ):
        raise ValueError("Sample size does not match the sum of male and female counts")
    # if not sample_size == sum((df[COL_DIAGNOSIS] == TERMURL_HEALTHY_CONTROL)) + sum(
    #     (df[COL_DIAGNOSIS] != TERMURL_HEALTHY_CONTROL)
    # ):
    #     raise ValueError(
    #         "Sample size does not match the sum of healthy controls and cases"
    #     )

    return str(sample_size), age, male, female  # , healthy_control, case


table = Table()
for col in ["", "N", "Age (years)", "Male", "Female"]:  # , "Control", "Case"]:
    table.add_column(col)

for task, df in zip(
    DATASETS_PER_TASK.keys(),
    [df_age, df_cog_decline, df_diag],
):

    table.add_row(task)

    for dataset in DATASETS_PER_TASK[task]:

        idx = df.query("dataset == @dataset")[
            ["participant_id", "session_id"]
        ].to_records(index=False)
        idx = list(map(tuple, idx))

        table.add_row(
            DATASET_MAP[dataset],
            *get_row_data(df_harmonized[dataset].loc[idx]),
        )

    # break

table
```
